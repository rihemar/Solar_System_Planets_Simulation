```python
import cv2
import mediapipe as mp
from mediapipe.tasks import python
from mediapipe.tasks.python import vision
from flask import Flask, jsonify
import threading

app = Flask(__name__)
latest_command = None

# --- MediaPipe Gesture Recognizer ---
model_path = r'gesture_recognizer.task'
base_options = python.BaseOptions(model_asset_path=model_path)
options = vision.GestureRecognizerOptions(base_options=base_options)
recognizer = vision.GestureRecognizer.create_from_options(options)

# Store previous index finger tip position to detect movement
prev_tip = None

def capture_gesture():
    global latest_command, prev_tip
    cap = cv2.VideoCapture(0)
    if not cap.isOpened():
        print("Error: Could not open webcam.")
        return

    while True:
        ret, frame = cap.read()
        if not ret:
            break

        rgb_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        mp_image = mp.Image(image_format=mp.ImageFormat.SRGB, data=rgb_frame)

        recognition_result = recognizer.recognize(mp_image)
        command = None

        if recognition_result.gestures:
            top_gesture = recognition_result.gestures[0][0]  # Most confident gesture

            if top_gesture.category_name.lower() == "closed_palm":
                command = "closed"

            elif top_gesture.category_name.lower() == "pointing_up":
                if recognition_result.hand_landmarks:
                    index_tip = recognition_result.hand_landmarks[0][8]  # Index finger tip
                    if prev_tip:
                        dx = index_tip.x - prev_tip[0]
                        dy = index_tip.y - prev_tip[1]

                        if abs(dx) > abs(dy):
                            if dx > 0.01:
                                command = "go_right"
                            elif dx < -0.01:
                                command = "go_left"
                        else:
                            if dy > 0.01:
                                command = "go_down"
                            elif dy < -0.01:
                                command = "go_up"

                    prev_tip = (index_tip.x, index_tip.y)

        if command:
            latest_command = command

@app.route('/gesture')
def get_gesture():
    return jsonify({"command": latest_command})

if __name__ == '__main__':
    thread = threading.Thread(target=capture_gesture, daemon=True)
    thread.start()
    app.run(host='0.0.0.0', port=5000)
```
